# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return (1 * (x > 0))
```

```python
x = plt.linspace(-10,10,100)
```

```python
plt.plot(x, ReLU(x), 'r', label='ReLU')
plt.plot(x, dReLU(x), 'b', label='dReLU')
plt.grid()
plt.title('ReLU/dReLU')
plt.legend(loc='upper left')
plt.text(0, 9, r'$f(x)= (abs(x) * (x > 0))$', fontsize=12)
plt.show()
```
